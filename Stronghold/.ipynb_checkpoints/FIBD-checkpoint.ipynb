### Mortal Fibonacci Rabbits

Given: Positive integers n≤100 and m≤20.

Return: The total number of pairs of rabbits that will remain after the n-th month if all rabbits live for m months.

```python
def calculate_mortal_fibon_rabbits(filename):
    file = open(filename,'rU')
    file = file.readline().split(' ')
    #print file
    n = int(file[0])
    #n = 6                           # !!!!! Value of the example, run the program with this to understand the working
    #print n

    m = int(file[1])
    #m = 3                           # !!!!! Value of the example, run the program with this to understand the working
    #print m
    
    k = 1                            #Fixed fo this exercise


    n_iterator = range(2,n+1)
    #print n_iterator
    a = 0
    b = 1
    
    mortal = [1]

    for i in n_iterator:
        if i < int(m+1):
            print i,'°'
            equa = a*(k+1) + b     #Where a = sexually mature rabbits, b = sexually immature rabbits and k is the litter (in pairs of rabbits)
            B = b
            A = a
            
            b = A * k 
            a += B  
            mortal.append(b)
            print 'Mortal: ',mortal
            print 'a: ',a
            print 'b: ',b
        
        elif i >= 4:
            print i,'°'            
            equa = a*(k+1) + b     #Where a = sexually mature rabbits, b = sexually immature rabbits and k is the litter (in pairs of rabbits)
            
            B = b
            A = a
            
            b = A * k
            a = a + B - mortal[i-(m+1)]
            mortal.append(b)
            print 'Mortal: ', mortal
            print 'Dead: ',mortal[i-(m+1)]
            print 'a: ',a
            print 'b: ',b
            print 'Result: ', a + b

```
