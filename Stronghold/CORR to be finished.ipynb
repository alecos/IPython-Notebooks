### Error Correction in Reads

**Given:** A collection of up to 1000 reads of equal length (at most 50 bp) in FASTA format. Some of these reads were generated with a single-nucleotide error. For each read s in the dataset, one of the following applies:

- s was correctly sequenced and appears in the dataset at least twice (possibly as a reverse complement);

- s is incorrect, it appears in the dataset exactly once, and its Hamming distance is 1 with respect to exactly one correct read in the dataset (or its reverse complement).


**Return:** A list of all corrections in the form "[old read]->[new read]". (Each correction must be a single symbol substitution, and you may return the corrections in any order.)



Sample Dataset


Sample Output


```python
def calculate_hamming_distance(seq1,seq2):
    
    hamming = 0
    
    if len(seq1) > len(seq2):
        
        hamming += len(seq1) - len(seq2)
        
        for i in range(0, len(seq2)):
            if seq1[i] != seq2[i]:
                hamming += 1
    
        return hamming
    
        
    elif len(seq1) < len(seq2):
        
        hamming += len(seq2) - len(seq1)
    
        for i in range(0, len(seq1)):
            if seq1[i] != seq2[i]:
                hamming += 1
         
        return hamming
    
    
    else:
        for i in range(0, len(seq1)):
            if seq1[i] != seq2[i]:
                hamming += 1
        
        return hamming
    
```

```python
from Bio import SeqIO
from Bio.Seq import Seq

file=open('files/rosalind_corr5.txt','rU')
fasta = list(SeqIO.parse(file, "fasta"))

seqs = []
raw_seqs = []
rev_raw_seqs = []
correct_seq = []
correct_rev_seq = []


### All sequences in raw data and calculating all the reverse complements

for i in range(0,len(fasta)):
    #print fasta[i].seq
    raw_seqs.append(str(fasta[i].seq))
    rev_raw_seqs.append(str(fasta[i].seq.reverse_complement()))                #calculating all the reverse complements

    
    
### Making a copy of raw data to later manipulate it:
###  - raw_seqs represents all the sequences
###  - seqs will be used as pool from wich subtract correct sequences and to store not already parsed sequences

for i in raw_seqs:
    seqs.append(i)
    

### Removing all correct sequences: if the read appears in the dataset at least twice (possibly as a reverse complement)

for i in raw_seqs:
    if raw_seqs.count(i) > 1:
        correct_seq.append(i)   #Store the sequence as correct
        seqs.remove(i)          #Remove it from the pool
    
    elif i in rev_raw_seqs:     #Check if the sequence was present as reverse complement in the pool
        correct_seq.append(i)   # ... if so it is added to the correct sequences and then removed from the pool
        seqs.remove(i)
        

        
### Calculating the reverse complement of every correct sequence

for i in correct_seq:
    correct_rev_seq.append(str(Seq(i).reverse_complement()))

    
    
### Checking wrong sequences:

for corr in correct_seq:
    for i in seqs:
        if calculate_hamming_distance(i,corr) == 1:
            print i + '->' + corr
            seqs.remove(i)
            
            
for corr_rv in correct_rev_seq:
    for i in seqs:
        if calculate_hamming_distance(i,corr_rv) == 1:
            print i + '->' + corr_rv
            seqs.remove(i)

            



print seqs
```
